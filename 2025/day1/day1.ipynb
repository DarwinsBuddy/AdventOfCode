```python
from math import fmod
import logging
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(levelname)s - %(message)s')

def parse_rotations(raw: str) -> list[str]:
    return raw.split('\n')

def read_input(filename: str, parser) -> list[str]:
    with open(filename, 'r') as f:
        raw = f.read()
        rots = parser(raw)
        return rots

def rot(pos, r) -> (int, int):
    result = 0
    # if we cross 0 to the left
    if pos > 0 and r < 0 and abs(r) > abs(pos):
        result = 1
    # if we cross 0 to the right
    elif pos < 0 and r > 0 and abs(r) > abs(pos):
        result = 1
    pos += r
    # check if exactly 0
    if pos == 0:
        result += 1
    else:
        # check for overshoot
        result += int(pos / 100)
        pos = int(fmod(pos, 100)) # modulo preserving sign
    return pos, result

def f(pos,r):
    res = int(abs(r) / 100)
    r = int(fmod(r,100))
    if r < 0:
        if (abs(r) >= pos and pos != 0):
            res += 1
        pos += (100 + r)
        pos = pos % 100
    else:
        pos += r
    res += int(abs(pos) / 100)
    pos = pos % 100
    #if pos == 0:
    #    res += 1
    return (pos,res)

def rotate(rots: list[int], pos=50) -> int:
    result = 0
    for r in rots:
        logging.debug(f"position {pos} rotate {r}")
        (pos, res) = f(pos, r)
        result += res
        ## reduce rotations to dial
        #res1 = int(abs(r) / 100)
        #r = int(fmod(r, 100)) # modulo preserving sign
        ## rotate the reduced rotations
        #pos, res2 = rot(pos,r)
        logging.debug(f"position {pos} 0-cross {res}")
        #result += res1 + res2
        
    return result

def run(filename: str, loglevel=logging.DEBUG):
    logging.getLogger().setLevel(loglevel)
    rots = read_input(filename, parse_rotations)
    rots = [int(r.replace('L','-').replace('R','+')) for r in rots]
    # print(rots)
    result = rotate(rots)
    logging.info(f"RESULT: {filename} -> {result}")
```
